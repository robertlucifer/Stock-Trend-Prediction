## Importing the required package of the model

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as pdr
from datetime import date
```

### Importing data for the model to calculate the stock trend for required amount of date

```python
today =date.today()
start = '2010-01-01'
end =date.today()
```

Importing the required company's stock data

```python
df=pdr.DataReader("AAPL", 'stooq',start,end)
```

```python
df
```

```python
df=df.reset_index()
```

```python
df
```

```python
df=df.drop(['Date'],axis=1)
```

```python
df
```

Ploting the closing of the specific company

```python
plt.plot(df.Close)
```

```python
ma100= df.Close.rolling(100).mean()
ma100
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.Close)
plt.plot(ma100,'red')
```

```python
ma200=df.Close.rolling(200).mean()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.Close)
plt.plot(ma100,'red')
plt.plot(ma200,'green')
```
